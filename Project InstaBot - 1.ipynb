#### Question 1.  Login to your Instagram Handle
#### Submit with sample username and password

```python
def LogIn(username, password):
    try :
        ## locating username textbox and sending username
        user_name = wait.until(EC.presence_of_element_located((By.NAME,'username')))
        user_name.send_keys(username)
        
        ## locating password box and sending password
        pwd = driver.find_element('name', 'password')
        pwd.send_keys(password)
        
        ## locating login button 
        button = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="loginForm"]/div[1]/div[3]/button/div')))
        button.submit()
        
        ## Save Your Login Info? : Not Now
        pop = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="_ac8f"]/div')))
        pop.click()
        pop = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_a9-z"]/button[contains(text(), "Not Now")]')))
        pop.click()
    except TimeoutException :
        print ("Something went wrong! Try Again")
```

```python
## Login to your Instagram Handle
LogIn(username, password)
```

#### Question 2. Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
#### Note : Make sure to avoid printing hashtags

```python
def search(s):
    try:
        ## locating search bar and sending text 
        search_box = wait.until(EC.presence_of_element_located((By.XPATH, '//span[@aria-describedby=":r3:"]')))
        search_box.click()
        search_box = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'x1lugfcp')))
        search_box.send_keys(s)
        time.sleep(2)
        
        ## waiting till all searched is located
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'xocp1fn')))
        
        waiter = wait.until(EC.presence_of_element_located((By.XPATH, '//a[contains(@class, "x1a2a7pz") and contains(@class, "x1dm5mii")]/div/div/div/div[2]/div/div/div/span[not(contains(span, "#"))]')))
        
        names = []
        for i in driver.find_elements('xpath', '//a[contains(@class, "x1a2a7pz") and contains(@class, "x1dm5mii")]/div/div/div/div[2]/div/div/div/span[not(contains(span, "#"))]'):
            names.append(i.get_attribute('innerHTML'))
            
            
        time.sleep(5)
        ## clearing search bar
        driver.find_element('class name', '_aawn').click()
        driver.refresh()
        return names
    
    except TimeoutException :
        print('No Search Found!')
```

#### Question 3. Searching and Opening a profile using 
#### Open profile of “So Delhi” 

```python
def search_open_profile(s):
    try:
        driver.refresh()
        ## locating search bar and sending text 
        search_box = wait.until(EC.presence_of_element_located((By.XPATH, '//span[@aria-describedby=":r3:"]')))
        search_box.click()
        search_box = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'x1lugfcp')))
        search_box.send_keys(s)
        
        s = s.replace(" ", "").lower()
        ## locating searched result
        res = wait.until(EC.presence_of_element_located((By.XPATH, f'//a[@href="/{s}/"]')))
        driver.execute_script("arguments[0].click();", res)

        time.sleep(5)
    
    except TimeoutException :
        print('No Search Found!')
```

```python
## for search and open 'So Delhi' instagram handle
search_open_profile('So Delhi')
```

#### Question 4.  Follow/Unfollow given handle - 
#### 1. Open the Instagram Handle of “So Delhi”
#### 2. Start following it. Print a message if you are already following
#### 3. After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def follow():
    try :
        ## locating follow button
        btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aj1-')))
        ## checking for text
        if btn.text == 'Follow' : 
            btn.click()
            print('Started Following')
            time.sleep(3)
        
        else : 
            print('Already Following')

    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def unfollow():
    try :
        ## locating follow button
        btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aj1-')))
        ## checking for text
        if btn.text !='Follow' : 
            btn.click()
            time.sleep(2)
            ## locating popup window (when you click on Unfollow button)
            pop_up = wait.until(EC.presence_of_element_located((By.XPATH, '//span[text()="Unfollow"]')))
            pop_up.click()
            print('Unfollowed')
            time.sleep(3)
            
        else : 
            print('Already Unfollowed')

    except TimeoutException :
        print("Something Went Wrong!")
```

```python
## for search and open 'So Delhi' instagram handle
search_open_profile('So Delhi')
## for following this instgram handle
follow()
## for unfollowing this instgram handle
unfollow()
```

#### Question 5.  Like/Unlike posts
#### 1. Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
#### 2. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
def Like_Post():
    try :
        time.sleep(2)

        posts_loaded = 0
        while posts_loaded < 30:
            ## Scroll down to load more posts
            driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
            time.sleep(2)
            ## Count the number of posts loaded so far
            posts = driver.find_elements(By.CLASS_NAME, '_aagw')
            posts_loaded = len(posts)
            
        for i in range(len(posts)):
            if i >= 30:
                driver.execute_script('window.scrollTo(0, -10000);')
                print('Liked top 30 posts')
                break  # Exit the loop after liking 30 posts
            posts[i].click()
            time.sleep(2)
            ## locating like/unlike button
            like = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aamw')))
            st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
        
            if st == 'Like' :
                like.click()
                time.sleep(2)
            else :
                print('You have already LIKED Post Number :', i+1)
                time.sleep(2)
            ## locating cross button for closing post
            driver.find_element(By.XPATH, '//div[@class="x160vmok x10l6tqk x1eu8d0j x1vjfegm"]//div[@role="button"]').click()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def Unlike_Post():
    try :
        driver.refresh()
        time.sleep(2)

        posts_loaded = 0
        while posts_loaded < 30:
            ## Scroll down to load more posts
            driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
            time.sleep(2)
            ## Count the number of posts loaded so far
            posts = driver.find_elements(By.CLASS_NAME, '_aagw')
            posts_loaded = len(posts)
     
        for i in range(len(posts)):
            if i >= 30:
                driver.execute_script('window.scrollTo(0, -10000);')
                print('Unliked top 30 posts')
                break  # Exit the loop after unliking 30 posts
            posts[i].click()
            time.sleep(2)
            ## locating like/unlike button
            like = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aamw')))
            st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
        
            if st == 'Unlike' :
                like.click()
                time.sleep(2)
            else :
                print('You have already UNLIKED Post Number', i+1)
                time.sleep(2)
            ## locating cross button for closing post
            driver.find_element(By.XPATH, '//div[@class="x160vmok x10l6tqk x1eu8d0j x1vjfegm"]//div[@role="button"]').click()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
```

#### Question 6.  Extract list of followers
#### 1. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
#### 2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def Extract_Followers():
    try :
        ## locating followers button and click on it
        followers_btn = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_ac2a')))
        followers_btn[1].click()
        time.sleep(2)
        ## locating followers list
        frame = driver.find_element(By.XPATH, '//div[@class="_aano"]')
        followers_count = 0
        names = []

        while followers_count < 500:
            ## Scroll down to load more followers
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight", frame)
            time.sleep(1)

            ## Extracting usernames
            followers = driver.find_elements('xpath', '//span[@class="_aacl _aaco _aacw _aacx _aad7 _aade"]')
            for follower in followers[followers_count:]:
                names.append(follower.text)
                followers_count += 1

            ## If we can't scroll anymore, break the loop
            if len(followers) == followers_count:
                break

        return names
        
    except TimeoutException :
        print("Something Went Wrong!")
```

#### Print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def Following():
    try :
        ## locating following button and click on it
        followers_btn = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_ac2a')))
        followers_btn[2].click()
        time.sleep(2)
        ## locating followers list
        frame = driver.find_element(By.XPATH, '//div[@class="x1ja2u2z x1afcbsf x1a2a7pz x6ikm8r x10wlt62 x71s49j x6s0dn4 x78zum5 xdt5ytf xl56j7k x1n2onr6"]')
        ## scrolling untill all users are located
        for i in range(20):
            time.sleep(1)
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",frame)
        
        names = []
        ## extracting userdata
        following = driver.find_elements('xpath', '//span[@class="_aacl _aaco _aacw _aacx _aad7 _aade"]')
        ## extracting username
        for i in following :
                names.append(i.text.split('\n')[0])
            
        return names
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
## for search and open 'foodtalkindia' instagram handle
search_open_profile('foodtalkindia')
## Extracting followers using Extract_Followers() function
followers_of_foodind = Extract_Followers()
## casting into set
followers_of_foodind = set(followers_of_foodind)
```

```python
## taking intersection so s1 contains only that user who is followed by me
s1 = (followers_of_foodind).intersection(followed_by_me)
if len(s1) == 0:
    print('No such users found')
else:
    ## now extracting my followers
    my_follower = Extract_Followers()
    my_follower = set(my_follower)
    ## taking intersection with s1, so s2 contains only users that I am following them but they don’t follow me
    s2=s1.intersection(my_follower)
    if len(s2) == 0:
        print('No such users found')
    else:
        for user in s2:
            print(user)
```

#### Question 7. Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
#### 1. If You have already seen the story.
#### 2. Or The user has no story.
#### 3. Or View the story if not yet seen.

```python
def Check_Story():
    try:
        ## locating story or profile pic
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aarh")))
    
        ## check the Profile photo size to find out story is available or not
        height = driver.find_element('class name', '_aarh').get_attribute('height')
        if int(height) == 208:
            print("Already seen the story")
        else:
            print("Viewing the story")
            driver.execute_script("arguments[0].click();",story)
    except:
        print("No story is available to view")
```
